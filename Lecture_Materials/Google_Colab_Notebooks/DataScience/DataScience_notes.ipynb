# Data Science L2

### Data types

#### Quantitative data: numeric values

If continuous: 
- histogram (single variable)
    - No gaps between the bars
    - No. of bins should be considered carefully
- scatter plot (2 variables)
- etc.

If discrete:
- bar chart

#### Qualitative data: non-numeric

Categorical: labelled

Binary: e.g. yes/no

Ordinal: ordered

### Measures of central tendency: a single value that attempts to describe a set of data

e.g. mean, median & mode - Identical for normal distribution. When analysing data: use more than one.

Median preferred when data is skewed.

__Order statistics__: Statistical methods that depend on the ordering of the data only - not the values.
e.g. median. 

### Box plots & quartiles

__Interquartile range__ = Q3 (75th percentile) - Q1 (25th percentile)

Maximum (upper whisker) = 1.5x IQR

Minimum (lower whisker) = 0.5x IQR

Data beyond the max & min are classed as outliers

### Distributions

__Bernoulli distribution__: When there are 2 possible outcomes, e.g. one toss of a coin.

__Uniform distribution__: Outcomes equally likely, e.g. rolling a dice]

__Binomial distribution__: Taking multiple binary outcomes, e.g. n tosses of a coin.

__Gaussian (aka Normal) distribution__: 

Probability Distribution Fuction (PDF): What is the probability of a particular value

Cumulative Distribution Function (CDF): What is the probability below a particular value

### Masking part of a heatmap

```python
# Import data
movies = pd.read_csv('moviesData.csv')

# Initialise figure
plt.figure(figsize=(10,10))

# first, create an array of zeros that matches the heatmap dimensions
mask = np.zeros_like(movies.corr(), dtype=bool)

# setting a value to True in the mask will mask the corresponding pixel in the heatmap
mask[np.triu_indices_from(mask)] = True

# apply the mask using the mask kwarg
sns.heatmap(movies.corr(), vmin=-1, cmap='coolwarm', annot=True, mask=mask)

plt.show()
```
